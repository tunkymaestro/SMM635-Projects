# Import relevant libraries

```python
import json
import gzip
import pandas as pd 
import numpy as np
```

# Read data

```python
def parse(path):
    g = gzip.open(path, 'rb')
    for i in g:
        yield json.loads(i)
        
def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient = 'index')
```
